```python
#An attempt for machine translation by using word2vec models
#Here I have trained two separate word2vec models one each for english and spanish 
#The models have been trained from wikipedia pages 
# (only those pages which are available in both languages have been added to the corpus)
#
# Using word vector representations of both languages I expect to find relations for transformations from one vector space to another 
```

```python
from bs4 import BeautifulSoup
import urllib
import urllib.request
import re
import word2vec
```

```python
# data generation for training
r_all = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages').read()
soup_all = BeautifulSoup(r_all,'lxml')
links=[]
for link in soup_all.find_all('a'):
    #print(link)
    links.append(link.get('href'))
links=list(set(links))
links.remove(None)
links = [s for s in links if re.match(r'/wiki/',s)!=None]
```

```python
def get_data(link,lang):
    r = urllib.request.urlopen('https://'+lang+'.wikipedia.org'+link)
    soup = BeautifulSoup(r,'lxml')
    
    text = ''
    for para in soup.find_all('p'):
        j = re.sub(r'\[([0-9]+)\]','',para.get_text())
        j = re.sub(r'\[(a-z)*\]','',j)
        if len(j)>10:
            text = text+j
    return text
```

```python
text_en = ''
text_es = ''
for link in links:
    try:
        en = get_data(link,'en')
        es = get_data(link,'es')
        text_en = text_en + en
        text_es = text_es + es
    except:
        print(link)
```
